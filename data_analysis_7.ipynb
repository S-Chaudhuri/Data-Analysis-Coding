```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv('fuel.csv',sep='\t')
a.head()
```

```python
a = a.drop('Unnamed: 0',axis=1)
a.head()
```

```python
cs = list(a.columns)
cs.pop(3)
print(cs)
```

```python
for c in cs:
    a[c]=a[c].astype(float)
a.info()
```

```python
a.shape
```

```python
a = a.drop_duplicates()
```

```python
a.isnull()
```

```python
m = a.isnull().sum()
m.plot.bar()
```

```python
a.dropna(how='all',inplace=True)
print(a)
```

```python
a.ffill(inplace=True)
print(a)
```

```python
a.to_csv('fuel2.csv',header=None,index=None)
```
