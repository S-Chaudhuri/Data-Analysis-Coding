```python
import pandas as pd
import numpy as np
r = pd.read_csv('log.txt',sep=' ')
r
```

```python
r.isnull()
```

```python
import matplotlib.pyplot as plt
plt.plot(r['age'],r['bought_insurance'],'go')
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(r[['age']],r.bought_insurance,test_size = 0.8)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
X_test
```

```python
y_p = lr.predict(X_test)
lr.predict_proba(X_test)
```

```python
lr.score(X_test,Y_test)
```
